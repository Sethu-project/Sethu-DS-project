```python
from faker import Faker
import random
import csv

fake = Faker()

def generate_student_profile():
    student_id = f"S{random.randint(100, 200)}"
    gender = random.choice(['Male', 'Female', 'Others'])
    name = (
        fake.name_male() if gender == 'Male'
        else fake.name_female() if gender == 'Female'
        else fake.name()
    )
    age = random.randint(15, 25)
    email = fake.email()
    phone = fake.phone_number()
    enrollment_year = random.choice(range(2020, 2026))
    course_batch = str(enrollment_year)
    city = fake.city()
    course_duration = random.choice([3, 4, 5])
    graduation_year = enrollment_year + course_duration

    return {
        'Student Id No': student_id,
        'Student_Name': name,
        'Age': age,
        'Gender': gender,
        'Mobile': phone,
        'Enrollment_Year': enrollment_year,
        'Course_Batch': course_batch,
        'Graduation_Year': graduation_year,
        'Email id': email,
        'City': city
    }

# ðŸ”„ Generate 100 student profiles
profiles = [generate_student_profile() for _ in range(100)]

# ðŸ“¤ Export to CSV
with open('student_profiles.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=profiles[0].keys())
    writer.writeheader()
    writer.writerows(profiles)

print("CSV file 'student_profiles.csv' created successfully.")
```
